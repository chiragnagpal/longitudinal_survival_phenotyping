```python
def condition_creator(boolx, booly, bool_oper):
    if bool_oper == '>':
        cond = boolx > booly
    elif bool_oper == '>=':
        cond = boolx >= booly
    elif bool_oper == '<':
        cond = boolx < booly
    elif bool_oper == '<=':
        cond = boolx <= booly
    elif bool_oper == '!=':
        cond = boolx != booly
    else:
        cond = boolx == booly
        
    return cond
```

```python
def quantity_stat(dataframe, name, bool_val, bool_oper, odd = True, duplication = 0):
    stat = []
    if duplication == 0:
        col = dataframe[name]
        cond = condition_creator(col, bool_val, bool_oper)
        
        return col[cond].count()
        
    else:
        if odd:
            for i in range(duplication):
                col_name = name + ' ' + str(i+1)
                col = dataframe[col_name]
                cond = condition_creator(col, bool_val, bool_oper) 
                stat.append(col[cond].count())
        else:
            for i in range(duplication):
                i += i+2
                col_name = name + ' ' + str(i)
                col = lipid_dataset[col_name]
                cond = condition_creator(col, bool_val, bool_oper) 
                stat.append(col[cond].count())
            
    return stat
```

```python
def total_NonNaN(dataframe, name, odd = True, duplication = 0):
    total = []
    if duplication == 0:
        return dataframe[name].count()
    else:
        if odd:
            for i in range(duplication):
                cur_name = name + ' '+ str(i+1)
                total.append(dataframe[cur_name].count())
        else:
            for i in range(duplication):
                i += i+2
                cur_name = name + ' ' + str(i)
                total.append(dataframe[cur_name].count())
            
    return total
    
```

```python
def yearly_percentage(data_list, tot_list):
    if len(data_list) is not len(tot_list):
        return -1
    percentages = []
    for i in range(len(data_list)):
        percentage = (data_list[i]/tot_list[i]) * 100
        percentages.append(round(percentage))
        
    return percentages
```

```python
from scipy.stats.stats import pearsonr

#Performs an annual correlation on a list of annual data featuring all patient values which is incorrect
def ann_corr(dataframe, featx, featy, years):
    corr_list = []
    for year in years:
        namex = featx + ' ' + str(year)
        namey = featy + ' ' + str(year)
        
        x = dataframe[namex]
        y = dataframe[namey]
        
        data = {'X': x,
               'Y': y}
        
        df = pd.DataFrame(data)
        df = df.dropna(axis=0)
        
        new_x = df['X']
        new_y = df['Y']
        
        corr_list.append(pearsonr(new_x, new_y))
        
    return corr_list
```

```python
def pre_proc(dataframe, featX, featY, years):
    if dataframe.empty:
        return pd.DataFrame()
    else:
        x_vals = []
        y_vals = []
        x_names = []
        y_names = []
        df = pd.DataFrame()
        
        for year in reversed(years):
            x_names.append(featX + ' ' + str(year))
            y_names.append(featY + ' ' + str(year))
            
        df = subdataset_extraction(x_names + y_names, dataframe)
              
        df = df.dropna(axis=0)
        df = df.reset_index()
            
    return df
```

```python
from scipy import signal
def cross_corr_2D(dataframe, featX, featY, years):
    df = pre_proc(dataframe, featX, featY, years)
    last_feats = [0]
    for i in range(len(years)):
        last_feats.append(i+len(years)+1)
    
    x = df.iloc[:,:-len(years)]
    y = df.iloc[:,last_feats]
    
    corr = signal.correlate2d(x, y)
    
    return corr
    
```

```python
import matplotlib.pyplot as plt

def avg_ann_metric(dataframe, featx, featy, years):
    """This function is meant to aid in obtaining the average
    metric (i.e. Systolic Blood Pressure, Diastolic Blood Pressure,...)
    value experienced by all patients annually or biannually"""
    
    # Check if dataframe is full
    if dataframe.empty:
        return -1
    x_avg = []
    y_avg = []
    
    for year in years:
        namex = featx + ' ' + str(year)
        namey = featy + ' ' + str(year)
        
        x = dataframe[namex]
        y = dataframe[namey]
        
        # Drop the NaN values which exist across both features
        data = {'Feature X': x,
               'Feature Y': y}
        
        df = pd.DataFrame(data)
        df = df.dropna(axis=0)
        df = df.reset_index()
        tot_patients = len(df.index)
        
        new_x = df['Feature X']
        new_y = df['Feature Y']
        
        new_x_tot = 0
        new_y_tot = 0
        
        for patient in range(tot_patients):
            new_x_tot += new_x[patient]
            new_y_tot += new_y[patient]
            
        x_avg.append(new_x_tot/tot_patients)
        y_avg.append(new_y_tot/tot_patients)
        
    # Plotting the Data
    plt.plot(years, x_avg, label='X')
    plt.plot(years, y_avg, label='Y')
        
    plt.xlabel('Years')
        
    plt.legend()

    correlation = pearsonr(x_avg, y_avg)
    print(x_avg)
    print(y_avg)
    print(correlation)
        
    return x_avg, y_avg, correlation
    
```
