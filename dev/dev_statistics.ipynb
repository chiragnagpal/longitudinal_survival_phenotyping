```python
def subdataset_extraction(columns, org_dataset):
    
    extracted_dataset = pd.DataFrame()
    
    for column in columns:
        temp_dataset = org_dataset.loc[:,[column]]
        extracted_dataset = pd.concat([temp_dataset, extracted_dataset], axis=1)
    return extracted_dataset
```

```python
def condition_creator(boolx, booly, bool_oper):
    if bool_oper == '>':
        cond = boolx > booly
    elif bool_oper == '>=':
        cond = boolx >= booly
    elif bool_oper == '<':
        cond = boolx < booly
    elif bool_oper == '<=':
        cond = boolx <= booly
    elif bool_oper == '!=':
        cond = boolx != booly
    else:
        cond = boolx == booly
        
    return cond
```

```python
def quantity_stat(name, bool_val, bool_oper, duplication = 0):
    stat = []
    if duplication is 0:
        col = dataset[name]
        cond = condition_creator(col, bool_val, bool_oper)
        
        return col[cond].count()
        
    else:
        for i in range(duplication):
            col_name = name + ' ' + str(i+1)
            col = dataset[col_name]
            cond = condition_creator(col, bool_val, bool_oper) 
            stat.append(col[cond].count())
            
    return stat
```

```python
def total_NonNaN(name, duplication = 0):
    total = []
    if duplication is 0:
        return dataset[name].count()
    else:
        for i in range(duplication):
            cur_name = name + ' '+ str(i+1)
            total.append(dataset[cur_name].count())
            
    return total
    
```

```python
def yearly_percentage(data_list, tot_list):
    if len(data_list) is not len(tot_list):
        return -1
    percentages = []
    for i in range(len(data_list)):
        percentage = (data_list[i]/tot_list[i]) * 100
        percentages.append(round(percentage))
        
    return percentages
```

## Number of Patients on Multiple Antihypertensive Medications Per Year
